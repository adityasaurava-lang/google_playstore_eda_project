```python
def size_process(items):
    if str(items)[-1]=='M':
        result=float(str(items).replace('M',''))
        result=result*1024
        return result
    elif str(items)=='k':
        result=float(str(items).replace('k',''))
        return result
    else:
        return str(np.nan)
        
```

```python
df_copy['Size']=df_copy['Size'].apply(size_process)
```
